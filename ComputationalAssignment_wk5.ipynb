```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('hot_jupiters.csv', engine='python', skiprows=103, header=0)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['pl_eqt'], df['pl_radj'], color='blue', alpha=0.7,s=25)
plt.xlabel('Equilibrium Temperature [K]')
plt.ylabel('Radius [Jupiter Radii]')
plt.title('Hot Jupiter Radius vs Equilibrium Temperature')
plt.grid(True)
plt.show()
```

```python
x = df['pl_eqt']    
y = df['pl_radj']  
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', alpha=0.7, label='Planets')
m, b = np.polyfit(x, y, 1)  
y_fit = m*x + b
plt.plot(x, y_fit, color='red', linewidth=2, label='Trend line')
plt.xlabel('Equilibrium Temperature [K]')
plt.ylabel('Radius [Jupiter Radii]')
plt.title('Hot Jupiter Radius vs Equilibrium Temperature')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['st_mass'], df['pl_radj'], color='blue', alpha=0.7,s=25)
plt.xlabel('Host Star Mass [Stellar Masses]')
plt.ylabel('Radius [Jupiter Radii]')
plt.title('Hot Jupiter Radius vs Stellar Mass of Host Star')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['st_met'], df['pl_radj'], color='blue', alpha=0.7,s=25)
plt.xlabel('Host Star Metallicity [dex]')
plt.ylabel('Radius [Jupiter Radii]')
plt.title('Hot Jupiter Radius vs Stellar Metallicity')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['pl_orbper'], df['pl_radj'], color='blue', alpha=0.7,s=25)
plt.xlabel('Orbital period [days]')
plt.ylabel('Radius [Jupiter Radii]')
plt.title('Hot Jupiter Radius vs Orbital Period')
plt.grid(True)
plt.show()
```
